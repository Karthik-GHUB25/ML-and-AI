#4) In a scatter plot, what does the 'hue' parameter represent?

**The color of data points based on a categorical variable**

The shape of data points

The size of data points

The transparency of data points

What hue Does in a Scatter Plot:
In libraries like Seaborn, the hue parameter is used to:

- Assign different colors to data points based on the values of a categorical or numerical variable

- Visually separate groups or classes in your data


```python
#import seaborn as sns
#import matplotlib.pyplot as plt

# Sample dataset
tips = sns.load_dataset("tips")

# Scatter plot with hue
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Tips by Gender")
plt.show()
```

#9) What is the correct way of plotting violin plot?

sns.violin(data=data, y = 'price',x = 'waterfront')

All of the answers

**sns.violinplot(data=data, y = 'price',x= 'waterfront')**

plt.violinplot(data=data, y = 'price',x='waterfront')

- sns.violinplot() is the Seaborn function designed specifically for creating violin plots.

- It combines the features of a boxplot and a KDE (kernel density estimate) to show the distribution of the data.

```python
#import seaborn as sns
#import matplotlib.pyplot as plt

sns.violinplot(data=data, x='waterfront', y='price')
plt.title("Price Distribution by Waterfront Status")
plt.show()
```

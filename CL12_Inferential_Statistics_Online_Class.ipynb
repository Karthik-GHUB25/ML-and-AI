# Probability Distribution

A probability distribution is a mathematical function that describes how likely different outcomes are for a random variable. It tells you:
- What values the variable can take
- How probable each value is

A probability distribution is just a way to show all the possible outcomes of something uncertain—and how likely each one is.

Think of it like a weather forecast:
- It doesn’t say exactly what will happen.
- It tells you the chances:
“There’s a 70% chance of rain, 30% chance of sunshine.”
That’s a probability distribution—just applied to weather.

There are two type of Probability Distribution

a. Discrete Probability distribution

b. Continuous probability distribution

***Discrete Probability distribution***

Exprement:
Tossing two coins on of the things will happen, Head & Head(HH), Head & Tail (HT), Tail & Head (TH), and Tail & Tail(TT). One of these four outcome can happen and nothing outside this can happen.

Below is the Sample Space

SS = {HH, HT, TH, TT}

Here the event is probability of getting at least one tail

Event: Getting at least one tail.

Outcome: Favorable outcome is { HT, TH, TT} = 3/4


Lets take the Sample Space

SS = {HH, HT, TH, TT},

the probability of getting HH is 1/4, HT is 1/4, TH is 1/4, and TT is 1/4 and this is called the probability distribution

SS = {HH, HT, TH, TT}

      1/4 1/4 1/4 1/4

When you combine the sample space and the probability of occurance is called probability distribution

In this case, each outcome is equally likely (fair coins), so the distribution is:

This is a uniform discrete probability distribution—because:
- The outcomes are discrete (countable)
- Each has equal probability


```python
sample_space = ['HH', 'HT', 'TH', 'TT']  # - This defines the sample space—all possible outcomes when tossing two coins
probability_distribution = {outcome: 1/4 for outcome in sample_space} #- This creates a dictionary where each outcome is assigned a probability of 1/4.
                                                                      #- Since the coins are fair, each of the 4 outcomes is equally likely.
                                                                      #- This is your uniform discrete probability distribution.
print(probability_distribution)
```

```python
favorable = ['HT', 'TH', 'TT']  # - These are the favorable outcomes for the event “getting at least one tail.”
prob_event = sum(probability_distribution[o] for o in favorable)  # - This calculates the probability of the event by summing the probabilities of the favorable outcomes.
                                                                  # - This line adds up the probabilities of the favorable outcomes.
                                                                  # - It calculates: P(\text{HT}) + P(\text{TH}) + P(\text{TT}) = 0.25 + 0.25 + 0.25 = 0.75
print(prob_event)
```

Drawing the probability distribution graph

```python
import matplotlib.pyplot as plt

# Step 1: Define the sample space and probabilities
sample_space = ['HH', 'HT', 'TH', 'TT']
probabilities = [0.25, 0.25, 0.25, 0.25]  # Uniform distribution

# Step 2: Plot the probability distribution
plt.figure(figsize=(6, 4))
plt.bar(sample_space, probabilities, color='skyblue', edgecolor='black')

# Step 3: Add labels and title
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Distribution of Tossing Two Fair Coins')
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 4: Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from statistics import mean, median, mode

salaries = [
    45000, 50000, 55000, 60000, 62000, 64000, 65000, 65000, 65000,
    66000, 68000, 70000, 75000, 80000, 85000
]

# Calculate central tendency measures
mean_val = mean(salaries)
median_val = median(salaries)
try:
    mode_val = mode(salaries)
except:
    mode_val = min(salaries)  # fallback if no unique mode

# Set plot style
sns.set(style="whitegrid")

# Create KDE plot
plt.figure(figsize=(10, 6))
sns.kdeplot(salaries, fill=True, color='teal', linewidth=2)

# Add vertical lines for mean, median, and mode
plt.axvline(mean_val, color='red', linestyle='--', linewidth=2, label=f'Mean: {mean_val:,.0f}')
plt.axvline(median_val, color='green', linestyle='-.', linewidth=2, label=f'Median: {median_val:,.0f}')
plt.axvline(mode_val, color='blue', linestyle=':', linewidth=2, label=f'Mode: {mode_val:,.0f}')

# Add labels and title
plt.title('Salary Distribution (Normal <No Skew>) with Mean, Median, and Mode', fontsize=14)
plt.xlabel('Salary')
plt.ylabel('Density')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate x values and corresponding y values for standard normal distribution
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

# Create the plot
plt.figure(figsize=(12, 6))
plt.plot(x, y, color='navy', linewidth=2)
plt.title('Standard Normal Distribution (68–95–99.7 Rule)', fontsize=14)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')

# Function to shade and label regions
def shade_and_label(start, end, color, label, xpos):
    x_fill = np.linspace(start, end, 500)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, color=color, alpha=0.5)
    plt.text(xpos, norm.pdf(xpos) + 0.01, label, ha='center', fontsize=10, color='black')

# Shade and label each region
shade_and_label(-1, 1, 'skyblue', '68.2%', 0)
shade_and_label(-2, -1, 'lightgreen', '13.6%', -1.5)
shade_and_label(1, 2, 'lightgreen', '13.6%', 1.5)
shade_and_label(-3, -2, 'salmon', '2.1%', -2.5)
shade_and_label(2, 3, 'salmon', '2.1%', 2.5)
shade_and_label(-4, -3, 'gray', '0.1%', -3.5)
shade_and_label(3, 4, 'gray', '0.1%', 3.5)

# Add vertical lines for reference
for z in range(-3, 4):
    plt.axvline(z, color='gray', linestyle='--', linewidth=1)

# Grid and layout
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# <mark>**Inferential Statistics **

Inferential statistics is the branch of statistics that allows us to draw conclusions about a population based on data from a sample. Instead of measuring every individual, we use probability theory to estimate, predict, or test hypotheses about the larger group.

Proof of Concept

We have Population and Sample.

Lets take the popular Indian Brand court case. Nestle Manufactures a product called Maggie Noodles.

Population is nothing but all the Maggie packets that are being produced and sold to the populaion. Nestle sells around 50,000,000 packets every year.

To study the packet it is impossible to collect all the packets that are being produced which 50 Million. It will take years and astronomical cost.

Logical way to study the packet is by taking a small sample packets from the population, for example 5000 packets

Whatever the conclusion that is derived from the 5000 packets will be applied on the 50 Million packets and this is called Inferential Statistics.

We need to know the poulation Mean value and again population is huge to calculate the mean, hence we calculate the Mean from the sample

Inferential Statistics

Tough the Inferential Statistics formula states Population Mean (PM) = Sample Mean (SM). In Reality it is not excatly the same.

PM = SM can be seen/called as Point Estimate

PM is mostly wrong so we have wrong so it requires some adjustment.

PM = SM ± Margin of Error

When you say 26 Months is the average life of a tyre, on the population is roughly around 26 months and not exactly 26 months.

lets 26 Months ± 1 Month

Formula to calculate Margin of Error (MOE)

PM = 26 - 1 month to 26 + 1 Month

PM = 25 Months to 27 Months. (Life of tyre ranges between 25 - 27 Month). Again it does not give any meaningful value. So it requires another change

This acutally is a Range or Interval. Instead of calling this formula as Population Mean, Making it as Confidence Interval (CI) makes more sense

CI = SM ± MOE

Population Mean is calculated using Sample Mean, It will be wrong to say Sample Mean is equal to Population Mean so Margin was error was added to Sample Mean. This resulted in Range rather than a single value, hence SM ± MOE becomes a Confidence Interval (CI).

90% Confidence = 26 Months ± 1 Month

However, you cannot give a huge range such as 0 - 100 Months.

Now lets put all these together

Formulkar for Confidence Interval

CI = x̅ ± (Z* (S/sqrt(n)))

x̅ = Sample Mean

S = Standard Deviation of the Sample

N = Sample Size (How many datapoint are in the sample)

Z* = Z-Score for a certain confidence level

(Z* (S/sqrt(n))) = Margin Of Error

```python
Z*        CL
1.65      90%
1.96      95%
2.58      99%
```

In real world you will be using the above three values to solve the problem. The above values are fixed.

```python
#9. Given a numpy array that has values = [1, 2, 3, 4 ,5]. Perform the following operation
#a. insert the element 6 at the end of the array
#b. insert the element 10 at the beginning of the array
#c. insert the element before 4
#d. update the element 5 as 50
#e. Delete the element 50 in the updated array
#f. create the function to search for the element 30
#g. Sort the array in ascending order

#9a. insert the element 6 at the end of the array
array12 = np.array([1, 2, 3, 4, 5])
array12 = np.append(array12, 6)
print(array12)
```

```python
#9b. Insert the element 10 at the beginning of the array
array12 = np.insert(array12, 0, 10)
print(array12)
```

```python
#9c. insert the element before 4
array12 = np.insert(array12, 3, 11)
print(array12)
```

```python
#9d. update the element 5 as 50
array12[array12 == 5] = 50
print(array12)
```

```python
#9e. Delete the element 50 in the updated array
array12 = np.delete(array12, np.where(array12 == 50))
print(array12)
```

```python
#9f. create the function to search for the element 30
def search_element(array, element):
    if element in array:
        return f"Element {element} found in the array."
    else:
        return f"Element {element} not found in the array."

search_element(array12, 30)
```

```python
#9g. Sort the array in ascending order
sorted_array_asc = np.sort(array12)
print(sorted_array_asc)
```

```python
#9g(a) Sort the array in descending order
sorted_array_desc = np.sort(array12)[::-1]
print(sorted_array_desc)
```

z-test:
Whenever there is percentages in the question, that means you have to use proportion over there.

Problem: A gorcery chains that claims 60% of the customers use reusable bags.

From a random sample of 300 customers, 165 used reusable bags

Check if the proportion differs from 0.60 (60%)

Step 1: Declare

Null Hypothesis (H₀): No evidence shows that usage rate differs from 0.60 (60%).

Alternative Hypothesis (H₁): Yes there is evidence shows that usage differs from 0.60 (60%)






```python
from statsmodels.stats.proportion import proportions_ztest

count = 165
nobs = 300 # Number of observation
p0 = 0.6

z_stat, pval = proportions_ztest(count = count, nobs = nobs, value = p0)

print(f"Z-statistic: {z_stat}")
print(f"P-value: {pval}")

if pval < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Problem: A coffee shop advertises that the average waiting time is 4 minutes.

A random sample of 12 customers gave these times (minutes):

[3.5, 4.2, 4.0, 4.4, 3.8, 4.1, 4.5, 3.7, 4.3, 3.9, 4.1, 4.0]

check if the actual mean differs from 4

Null Hypothesis (H₀): wait time is 4.

Alternative Hypothesis (H₁): There is no evidence shows wait time is 4 minutes. Wait time not equal to 4 minutes


```python
from scipy.stats import ttest_1samp

x = [3.5, 4.2, 4.0, 4.4, 3.8, 4.1, 4.5, 3.7, 4.3, 3.9, 4.1, 4.0]

t_stat, p_val = ttest_1samp(x, popmean = 4)

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Problem: A factory wants to check if two machines produce bolts with similar variability in length.

Machine A (mm): 10.1, 10.3, 9.9, 10.2, 10.0

Machine B (mm): 9.8, 10.5, 10.2, 9.9, 10.4

test at 5% signinificance whether the variances differ

- Null Hypothesis (H₀): The variances are equal

→ There is no significant difference in variability between Machine A and Machine B.

- Alternative Hypothesis (H₁): The variances are not equal

→ There is a significant difference in variability

f_test formula

F = S1^2/S2^2

Where Variance of S2 is always greater the S1

```python
from scipy.stats import f_oneway

Machine_A = [10.1, 10.3, 9.9, 10.2, 10.0]

Machine_B = [9.8, 10.5, 10.2, 9.9, 10.4]

f_stat, p_val = f_oneway(Machine_A, Machine_B)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

f_test

Lets take an example where you are selling bulbs. Bulbs are supplied from two different manufacturers. Bulbs are in two groups Group A and Group B. You have to decide which manufacturer provides good bulbs. Here you have to compare bulbs from Group A and Group B.

The suitable test here is f_test. It is used to compare the variance between two groups.

You use ANOVA (Analysis Of VAriance) when there is more than two groups.


ANOVA Formula (One-Way)

F-statistic:

F = \frac{\text{MS}_{\text{between}}}{\text{MS}_{\text{within}}}

Where:
- MS_between = Mean Square Between Groups
- MS_within = Mean Square Within Groups



Problem: Compare mean weekly sales across three store locations.

store_A = [52, 55, 50, 58, 53]

store_B = [49, 47, 50, 46, 48]

store_C = [60, 62, 59, 61, 63]

Here we can use ANOVA

```python
from scipy.stats import f_oneway

store_A = [52, 55, 50, 58, 53]
store_B = [49, 47, 50, 46, 48]
store_C = [60, 62, 59, 61, 63]

f_stat, p_val = f_oneway(store_A, store_B, store_C)

print(f"F-statistic: {f_stat}")
print(f"P-value: {p_val}")

if p_val < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

Problem: A supermarket surveyed three age groups of customers to check whether their preferred payment method (card or cash) is related to group

Age Group               Card        Cash
Young (18-30)             40          25
Middle-Aged(31 - 50)      35          30
Senior (51+)              20          15

Goal: Test whether payment method is independent of age group at the 5% significant level.


- Null Hypothesis (H₀): Payment method and age group are independent

→ Preference for card or cash does not depend on age group.

- Alternative Hypothesis (H₁): Payment method and age group are associated

→ Preference for payment method varies by age group.


Here we use Chi-Squared test.

Chi-Squared Test, whenever you want to study the association with two categorical variables. Example. Men like shoes and women like jewelries

```python
import numpy as np
from scipy.stats import chi2_contingency

observed = np.array([[40, 25], [35, 30], [20, 15]])

chi2_stat, p_val, dof, expected = chi2_contingency(observed)

print(f"Chi-Squared Statistic: {chi2_stat}")
print(f"P-value: {p_val}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

if p_val < 0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

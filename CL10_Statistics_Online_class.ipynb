What Is Statistics?

Statistics is the science of collecting, organizing, analyzing, and interpreting data to make informed decisions. It‚Äôs not just about numbers ‚Äî it‚Äôs about patterns, uncertainty, and storytelling through data. statistics is a branch of applied mathematics that focuses on collecting, analyzing, interpreting, and presenting data.
üîç Two Branches of Statistics:
Type	Purpose	Example
Descriptive	Summarizes data you already have	Average delivery time last month
Inferential	Makes predictions or decisions from samples	Predicting next month‚Äôs defect rate from a sample

The types of descriptive statistics include:
1. Distribution: This describes how data points are spread across various values or ranges within a dataset. Distribution refers to the frequency of each value in a dataset, often represented in a frequency distribution table or graph. It shows how often each value appears in the dataset
2. Measures of Central Tendency: These summarize a dataset by identifying its central point or typical value, such as mean, median, and mode.
3. Variability (or Dispersion): These measures how spread out the values are in a dataset, indicating the degree of variation among the data points.
These categories help in summarizing and understanding the characteristics of a dataset effectively. variability or dispersion describes how spread out the values are around the central tendency. It can be measured by various statistical measures such as range, interquartile range, variance, and standard deviation. Variability helps to understand the degree of spread in the data, which is crucial for making informed decisions and predictions based on the data.

Inferential Statistics
Inferential statistics is a branch of statistics that uses sample data to make generalizations, predictions, or inferences about a larger population. Unlike descriptive statistics, which summarize data, inferential statistics go beyond the data at hand to estimate parameters, test hypotheses, and predict future trends.
Inferential statistics can be defined as a field of statistics that uses analytical tools for drawing conclusions about a population by examining random samples.
Types of Inferential statistics
Hypothesis Testing
Regression Analysis

```python
# Manual way of finding the median.
# Below has the even total number of values
data = [5, 2, 7, 4, 1, 8, 9, 6]                                     #position  0  1  2  3  4  5  6  7
data.sort()         # After sorting [1, 2, 4, 5, 6, 7, 8, 9]                  [1, 2, 4, 5, 6, 7, 8, 9]
n = len(data)       # data length is 8
if n % 2 == 0:  # n % 2 is modulo operation.  It calculates the remainder. when n is divided by 2 and the remainder is 0 then n is even number.
    median = (data[n//2 - 1] + data[n//2]) / 2      # n // 2 is the floor division operator. It divides n by 2 and returns the largest integer less than or equal to the result
else:                                               # data[n//2 - 1] = data[3] = 5 and data[n//2] = data[4] = 6. (5+6)/2 = 5.5
    median = data[n//2]
print(median)                                       # 5.5
```

<mark> Variance

Variance measures how far each number in a dataset is from the mean (average) and thus from every other number in the set. It is denoted by the symbol \(œÉ\^2\) for population variance and \(s\^2\) for sample variance. A higher variance indicates that the data points are more spread out from the mean, while a lower variance suggests that they are closer to the mean

Small variance = the numbers are close to average

High Variance = The numbers are spread out

Variance is the average of the squared distance from mean

Below is the formula for Population

= (Œ£ ( x - Œº)^2) / N

x = data point (each value in the dataset)

Œº = Mean of the data

N - Number of values

data_point = [2, 4, 6]

X = 2, 4, 6

Œº = (2 + 4 + 6) / 3 = 12/3 = 4

now subtract the mean from each data point

( x - Œº)^2 = (2 - 4)^2  = 4

( x - Œº)^2 = (4 - 4)^2  = 0

( x - Œº)^2 = (6 - 4)^2  = 4

(Œ£ ( x - Œº)^2) / N = (4 + 0 + 4)/3 = 8/3 = 2.666 = 2.67

Variance = 2.67

===============================================================================

Below is the formula for Sample

= (Œ£ ( x - Œº)^2) / N-1

data_point = [2, 4, 6]

X = 2, 4, 6

Œº = (2 + 4 + 6) / 3 = 12/3 = 4

now subtract the mean from each data point

( x - Œº)^2 = (2 - 4)^2  = 4

( x - Œº)^2 = (4 - 4)^2  = 0

( x - Œº)^2 = (6 - 4)^2  = 4

(Œ£ ( x - Œº)^2) / N-1 = (4 + 0 + 4)/3-1 = 8/2 = 4

===============================================================================

<mark> Percentiles

Percentiles are a powerful way to understand how a value compares to the rest of a dataset.

A percentile tells you the relative standing of a value within a dataset. It answers the question:

‚ÄúWhat percentage of the data falls below this value?‚Äù

So if a value is at the 80th percentile, it means 80% of the data is less than or equal to it, and 20% is greater.

Percentile ‚Äì 25, 50, 75
These are specific percentiles that are commonly used because they help us to understand the data spread


Example

Lets say you score 80th percentile, that mean you scored better that 80 students.


```python
Student_Score = [55, 60, 65, 70, 75, 80, 85, 90, 95]      # Arr Index   0   1   2   3   4   5   6   7   8
# Calculate the percentile value                                  Arr [55, 60, 65, 70, 75, 80, 85, 90, 95]
percentile_value = np.percentile(Student_Score, 80)
print(percentile_value)
```

The above shows 87

What it means is 80% of the score is less than or equal to 87.

How it works?

Step 1: Compute the Rank Position

\text{Rank} = \frac{P}{100} \times (n - 1)  

or

Rank = (P/100) * (N-1)

Rank = (80/100) * (9-1) = 6.4

Where:

- P = 80
- n = 9 (number of values)

So, NumPy looks between the 6th and 7th values (indexing starts at 0):
- Index 6 ‚Üí 85
- Index 7 ‚Üí 90

Step 2: Interpolate Between These Values

\text{Percentile Value} = X_{\text{lower}} + f \times (X_{\text{upper}} - X_{\text{lower}})

\text{Percentile Value} = 85 + 0.4 \times (90 - 85) = 85 + 2 = 87

85+ 0.4 * (90-85) = 85 + 0.4 * 5 = 85 + 2 = 87

Note: we take the decimal part of the rank, which is 0.4, and use it to interpolate between 85 and 90: (i.e: 0.4 is taken from 6.4)

<Mark> Quartiles

Quartiles split a sorted dataset into four equal sections, each containing 25% of the data.

How does quartile divide data into 4 equal parts when the dataset size doesn‚Äôt divide cleanly by 4.

The Truth: Quartiles Don‚Äôt Require Perfect Division

Quartiles divide the range of values, not the number of data points, into four equal parts. So even if you have 9 data points, you can still compute Q1, Q2, and Q3 using interpolation.

        index =   0   1   2   3   4   5   6   7   8
                 55  60  65  70  75  80  85  90  95   

Student_Score = [55, 60, 65, 70, 75, 80, 85, 90, 95]

Sorted, with 9 values.

How Quartiles Are Calculated (Using Percentiles)

- Q1 = 25th percentile ‚Üí position = 0.25 \times (n - 1) = 0.25 \times 8 = 2.0

- So Q1 = value at index 2 ‚Üí 65

- Q2 = 50th percentile ‚Üí position = 0.5 \times 8 = 4.0
- Q2 = value at index 4 ‚Üí 75
- Q3 = 75th percentile ‚Üí position = 0.75 \times 8 = 6.0
- Q3 = value at index 6 ‚Üí 85

No interpolation needed here because the positions land exactly on whole-number indices.


```python
Student_Score = [55, 60, 65, 70, 75, 80, 85, 90, 95]

# import numpy as np

Q1 = np.percentile(Student_Score, 25)
Q2 = np.percentile(Student_Score, 50)
Q3 = np.percentile(Student_Score, 75)

print("Q1:", Q1)
print("Q2 (Median):", Q2)
print("Q3:", Q3)
```

```python
# Python Function: Quartile Calculator with Interpolation
# import numpy as np

def calculate_quartiles(data):
    data_sorted = sorted(data)
    n = len(data_sorted)

    def interpolate(percentile):
        pos = percentile / 100 * (n - 1)
        lower_index = int(np.floor(pos))
        upper_index = int(np.ceil(pos))
        fraction = pos - lower_index

        lower_value = data_sorted[lower_index]
        upper_value = data_sorted[upper_index]

        interpolated = lower_value + fraction * (upper_value - lower_value)
        return interpolated

    Q1 = interpolate(25)
    Q2 = interpolate(50)
    Q3 = interpolate(75)

    print("Q1:", Q1)
    print("Q2 (Median):", Q2)
    print("Q3:", Q3)

# Example usage
Student_Score = [55, 60, 65, 70, 75, 80, 90, 95]
calculate_quartiles(Student_Score)
```

What Is Linear Interpolation?

Linear interpolation assumes that the data between two known points increases at a constant rate ‚Äî a straight line. So when a quartile falls between two data points, we estimate its value by proportionally ‚Äúsliding‚Äù between them.

You calculated Q1 as falling at position 2.25 in the sorted list:

‚Ä¢ 	Index 2 = 3

‚Ä¢ 	Index 3 = 40

‚Ä¢ 	So Q1 lies 25% of the way between 3 and 40

Using linear interpolation:

Q1 = 3 + 0.25 \times (40 - 3) = 3 + 9.25 = \boxed{12.25}

This assumes a straight-line increase from 3 to 40 ‚Äî no jumps, no curves, just a smooth transition.

*   Null Hypothesis
*   Alternate Hypothesis

Alternate Hypothesis: Suggests there is an effect or difference. It is opposite of Null Hyposthesis. ‚ÄúThe new QA process reduces defect rate‚Äù

p_value = Probability value

The p-value is a number between 0 and 1 that tells you how likely your observed data would occur if the null hypothesis were true.

In simpler terms:
It measures the strength of evidence against the null hypothesis.

Example:

Suppose a company claims their bulb lasts 1000 hrs, upon testing on 30 sample bulbs the result showed the bulb lasts only 950 Hours.

if p_value = 0.02 there are only 2% chances we would see such a big difference if bulbs really last 1000 hours

evidence is strong -> reject H0 -> Claim is false. Accept Alternate Hypotheses

if p=0.40 -> Accepting the H0, the claim by the company is true.

tells you about the symmetry or asymmetry in a data distribution

Types of skewness

<mark> **Inferential Statistics**

Inferential Stats is a way to make guesses or conclusions about a big group (population) by looking at small part (sample).

Population - Population means the entire group or full data

Sample - Subset or small part of the data.

**Types of Sampling**

1. Random Sampling: Sample collected randomly. (choosing a name from a hat). Every member of the population has an equal chance of being selected.

2. Systematic Sampling: Sampling based on condition. Select every k-th item from a list after a random starting point. Ex. From a list of 1,000 products, you randomly start at item 7 and then select every 10th item (17, 27, 37‚Ä¶)

3. Stratified Sampling: Divide the population into subgroups (strata) based on shared characteristics, then randomly sample from each group. Ex. In a company with 60% engineers and 40% designers, you sample 60 engineers and 40 designers to reflect that ratio

4. Cluster Sampling: Grouping the data based on the similarity or you can say you are splitting the group based on similarity, and choose a cluster randomly.

example:

- Cluster A: Downtown
- Cluster B: Lakeside
- Cluster C: Suburban East

Each neighborhood is a cluster.
üîπ Step 2: Randomly select clusters
You randomly pick Cluster B and Cluster D.
üîπ Step 3: Survey everyone in those clusters
You go door-to-door in Lakeside and Rural West, asking every household about their favorite ice cream flavor

The result of the experiment is the out come.

*Event* :

What result are you interested in. Lets say you wanted head when flipping the coin. Two types of event

a. Simple event: one specific out come

b. compound event: There will be more than one out come.

A compound event in probability refers to any event that involves two or more simple events happening together‚Äîeither simultaneously or sequentially.

it has four types
b1. Independent Events: One event does not affect the other

    P(A \cap B) = P(A) \times P(B)

b2. Dependent Events: One event does affect the outcome of the other

    P(A \cap B) = P(A) \times P(B|A)

b3. Mutually Exclusive: Events cannot happen at the same time

    P(A \cup B) = P(A) + P(B)

b4: Mutually Inclusive: Events can happen together

    P(A \cup B) = P(A) + P(B) - P(A \cap B)

    probability of getting 3 in one dice is 1/6

P(\text{no 3 on both}) = P(\text{no 3 on die 1}) \times P(\text{no 3 on die 2}) = \frac{5}{6} \times \frac{5}{6} = \frac{25}{36}

P(No 3 on both) = P(No 3 on die 1) * P(No 3 on die 2)

P(No 3 on both) = 5/6 * 5/6 = 25/36

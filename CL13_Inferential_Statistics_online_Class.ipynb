# Central Limit Theorem

Core Idea: Regardless of the population’s distribution, the sampling distribution of the sample mean becomes approximately normal as the sample size grows (typically n ≥ 30).

Theorem are right as this was tested by various statistians

Imagine you're overseeing a production line for light bulbs. The lifespan of each bulb varies due to minor inconsistencies in materials and processes. The population distribution of lifespans is skewed, not normal.

Here's how CLT comes into play:
- You randomly select 30 bulbs from each batch and record their lifespans.
- You repeat this sampling process across 500 batches.
- For each batch, you compute the average lifespan.
- When you plot the distribution of these 500 sample means, it forms a bell-shaped curve, even though the original population of lifespans is skewed.

Discussion

Lets take a large population, From the large population lets take multiple small samples (500 Samples) where each sample has more than 30 data points.

To begin with lets keep the datapoint on each sample is almost same. Lets keep 35 datapoints on each sample.

Now lets take the Mean(Average of 35 datapoints) of each sample.

Here we have a large collection of Mean values (A total of 500). Now lets take all these sample Mean and plot these values using either SNS.KDEPLOT or SNS.DISPLOT. Whenever you have collection of values in float, that is your continuous data which can be used for making the plot (KDE or DISPLOT). If you have Integer than you make a bar chart.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate a skewed population (e.g., exponential distribution)
population = np.random.exponential(scale=1000, size=15000)

# Generate 500 sample means (each sample has 35 data points)
sample_means = [np.mean(np.random.choice(population, size=35, replace=False)) for _ in range(500)]

# Create a KDE + histogram plot using seaborn's displot
sns.displot(sample_means, kde=True, bins=35, color='steelblue', height=5, aspect=1.5)

plt.title('Distribution of 500 Sample Means (n=35)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.tight_layout()
plt.show()
```

What you will observe is, it will always come out to be Normal Distribution. In normal distribution you can find Empirical rule. The Empirical Rule—also known as the 68–95–99.7 Rule—is a simple yet powerful way to understand how data behaves in a normal distribution (bell curve).

We can also apply Z-Score to find the area behind a particular number or between two numbers.

So the theorem says

When you have a population, and you take multiple sample from it, and Sample mean from each sample will be a Normal Distribution.

Note: Always make sure each sample should have more than 30 data Points.

# CLT and Small population, can it work?

Great question, Karthik—and it shows you're thinking critically about the **limits and assumptions** behind the Central Limit Theorem (CLT). The answer depends less on the *population size* and more on the *sampling conditions* and *sample size relative to the population*.

---

🧠 CLT and Small Populations: Can It Work?

Yes, you **can** apply CLT to small populations—but only if certain conditions are met:

✅ **Key Conditions for CLT to Hold**
1. **Random Sampling**: Every member of the population must have an equal chance of being selected.
2. **Independence**: Each sample must be independent of the others.
3. **10% Rule**: If sampling *without replacement*, your sample size should be **no more than 10% of the population** to preserve independence.
4. **Large Sample Size**: Typically, **n ≥ 30** is considered “large enough” for CLT to kick in—even if the population is skewed.

---

📉 Example: Small Population Scenario

Let’s say your population is **500 people**. According to the 10% rule:
- You should keep your sample size **≤ 50**.
- If you draw **many samples of size 30**, CLT can still apply to the **distribution of sample means**.

So technically, you could apply CLT to a population as small as **300**, **500**, or even **100**, as long as:
- Your sample size is small relative to the population.
- You draw multiple samples and analyze the distribution of their means.

---

🔍 Bottom Line
There’s no hard minimum population size for CLT—but the **sample size must be large enough**, and **sampling must preserve independence**. If those hold, CLT can work even on small populations.

Want to simulate this with a population of 100 or 300 and visualize how the sample means behave? That would be a great way to test the boundaries of CLT in your portfolio.

What Is a p-value?
The p-value is the probability of observing your sample data—or something more extreme—if the null hypothesis (H₀) were true.
In simpler terms:

Key Points
• 	Range: Always between 0 and 1
• 	Low p-value (e.g., < 0.05): Strong evidence against the null hypothesis → you reject H₀
• 	High p-value (e.g., > 0.05): Weak evidence against H₀ → you fail to reject H₀

p-value is the probability of getting such type of evidences or even more extreme evidence than what we got, under the assumption that 'Null hypothesis is true"

p-value refers to how strong evidence we have?

Summary

• 	As evidence against the null hypothesis (H₀) increases, the p-value decreases.
• 	If the evidence is weak or minimal, the p-value remains high.

Interpretation Guidelines
• 	Low p-value (e.g., < 0.05) → Strong evidence against H₀ → Reject H₀
• 	High p-value (e.g., > 0.05) → Weak evidence against H₀ → Fail to reject H₀

A high p-value does not prove H₀ is true—it simply means the data is not surprising under H₀, so we don’t have enough evidence to reject it.

Think of the p-value as a measure of how “surprised” you should be by the data if H₀ were true.

• 	Very surprised? → Low p-value → Reject H₀

• 	Not surprised? → High p-value → Fail to reject H₀

**Significance Level (α)**

The significance level, denoted by alpha (α), is the threshold you set to decide whether to reject the null hypothesis (H₀). It represents the maximum probability you're willing to accept for making a Type I error—that is, rejecting H₀ when it’s actually true.

The Maximum amount of error you are ready to make when you reject a Null Hypothesis.

SL = 0.05 = 5%

### Z-Test

Z-Test has two versions

1. One-Sample Z-Test: Used to determine whether the mean of a single Sample is different from a known population Mean.

2. Two-Sample Z-Test: Used to compare the means of two independent samples to see if they are significantly different from each other

Assumptions

For a z-test to yield valid/correct results, these assumptions must be met:

*Normal Distribution:* The data should be approximately normally distributed. This assumption is satisfied with large sample sizes due to the Central Limit Theorem.

One way to prove this is to make a plot and see if this is normally distributed.

According to CLT, if you have a sample size greater than 30 you can say it is normally distributed

**### One-Sample Z-Test:** Used to determine whether the mean of a single Sample is different from a known population Mean.

Business Problem Statement:

A retail chain claims that the average monthly spending of its loyality program members is Rs.5000. The management wants to verify this claim using a random sample of 50 members. Population standard deviation is Rs.600

Hypotheses

Null Hypothesis (H₀): The average monthly spending is Rs.5000 (μ = 5000)

Do not reject NH [x̄ (Sample Mean) = μ (Population Mean)]

Alternative Hypothesis (H₁): The average monthly spending is not Rs.5000
(μ != 5000)

reject NH [x̄ (Sample Mean) != μ (Population Mean)]

In Alternative Hypothesis (H₁),
*   If you are having '<>' symbol it will be two tailed test, Which means rejection reason is only on both sides
*   If you are having '<' or '>' symbol it will be one tailed test. Which means rejection reason is only on one side


Business Scenario
*   Claimed mean (μ) : Rs.5000
*   Sample Mean (x̄): Rs.5200 (supposedly)
*   Population Std Dev (σ): Rs.600
*   Sample Size (n): 50

```python
import numpy as np
import math
from scipy.stats import norm

sample_data = [
    5105.42, 5230.87, 4980.16, 5345.33, 5195.68, 5070.91, 5255.24, 5120.39, 5300.77, 4955.12,
    5180.66, 5275.48, 5035.29, 5150.73, 5405.84, 4990.36, 5210.15, 5085.92, 5240.61, 4930.88,
    5165.47, 5320.04, 5005.38, 5295.96, 5140.22, 4965.75, 5225.63, 5050.19, 5190.81, 5350.26,
    4975.44, 5265.59, 5100.03, 5315.91, 5020.67, 5205.18, 5060.86, 5280.34, 5000.49, 5385.72,
    4945.11, 5175.65, 5095.37, 5330.25, 5015.93, 5235.06, 5040.78, 5360.49, 4985.20, 5305.88
]

# The above data shows how much of these customers spent last month
pop_mean_μ = 5000
sample_mean_x̄ = np.mean(sample_data)
Pop_Std_Dev_σ = 600
Sample_Size_n = 50

Z = (sample_mean_x̄ - pop_mean_μ) / (Pop_Std_Dev_σ / math.sqrt(Sample_Size_n))
print(f"Sample Mean: {sample_mean_x̄:.2f}")
print(f"Z-score: {Z:.2f}")
```

In the definition it is mentioned

Alternative Hypothesis (H₁): The average monthly spending is not Rs.5000 (μ != 5000)

This means you have rejection on both sides

Acceptance area is 95% which means it stretches from -1.96 to +1.96
Here the z value of sample mean is 1.86

which falls with in -1.96 to +1.96

Final Conclusion

Since Z = 1.86 falls within the acceptance region (between -1.96 and +1.96), you fail to reject the null hypothesis.

Interpretation:

There is not enough statistical evidence to conclude that the average monthly spending of loyalty members is different from ₹5000. The observed sample mean of ₹5157.51 could reasonably occur due to sampling variability.

## Two Sample T-Test

Compare the means of two independent groups to see if the difference is statistically significant

Key Assumptions
- Independence: Samples are unrelated (e.g., Group A ≠ Group B).
- Normality: Data in each group is approximately normally distributed.
- Equal Variance (optional): Use Student’s t-test if variances are equal, or Welch’s t-test if not.

When to Use a Two-Sample T-Test
You use it when:
- You have two separate groups (e.g., customers from Region A vs Region B).
- You're measuring the same variable across both groups (e.g., average payment delay, recovery yield, or satisfaction score).
- You want to test whether the difference in means is due to chance or reflects a real effect

To decide we compare

SM1 vs SM2

SM1-SM2 = 0

Business Problem Statement

An HR department wants to compare the average monthly salaries of employees in two different departments (A and B) to determine if there is a significant difference.

Hypotheses

Null Hypothesis (H₀): The mean salary in Department A equals that in Department B (μ1=μ2).

Alternative Hypothesis (H₁): The mean salaries are different (μ1≠μ2).

Before performing a two-sample t-test, it is important to check two key assumptions:

Normality: Each sample should be approximately normally distributed.

Equality of Variances: The two samples should have similar variances (for the standard t-test). This means spread of the data should be the same. For example currency used in both department should be the same.

So your decision logic should be:
• 	If p-value < α (e.g., 0.05) → Reject H₀ → Significant difference.
• 	If p-value ≥ α → Fail to reject H₀ → No significant difference

```python
import numpy as np

# Monthly salaries/month in USD for two departments
dept_a = np.array([4500, 4700, 4200, 4800, 4600, 4400, 4550, 4650, 4750, 4300]) # Data science Department
dept_b = np.array([4000, 4150, 4100, 3950, 4200, 4050, 4100, 4000, 4150, 4050]) # Engineering Department

# Two Key Assumptions
# Assumption 1: Use Shapiro-Wilk to test for normality of both the data
from scipy.stats import shapiro
Stat_a, p_value_a = shapiro(dept_a)
Stat_b, p_value_b = shapiro(dept_b)
print("P-value of Department A: ", {p_value_a}, "P-value of Department B: ", {p_value_b})

# Assumption 2: Equality of Variances
from scipy.stats import levene

# use Levene's test for equal variances
stat_levene, p_levene = levene(dept_a, dept_b)

print(f"Levene's test p-value = {p_levene:.4f}")

if p_levene > 0.05:
    print("Variances are equal.")
    equal_var = True
else:
    print("Variances are NOT equal.")
    equal_var = False

# Run two sample t-test
from scipy.stats import ttest_ind  # test_ind is the function used for running two Sample t-test
                                   # Two department are independent to each other

# Two-sample t-test (Welch's if variances unequal)
t_stat, p_value = ttest_ind(dept_a, dept_b, equal_var=False) # This gives two outputs T-statistic & P-value
                                                             # First it calculates T-statistic and using this value it calculates P-Value
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")                             # Compare the p_vlaue with significance level and decide whether to reject or keep Null Hypothesis
```

If the p_value of Shapiro-Wilk test is greater than 0.05, the sample can be considered normally distributed

Here the p_value of both departments are greater than 0.05, hence they are normally distributed.

If the data is not normally distributed then you may want to increase the data or simply take more than 30 data.

Levene's test p-value = 0.0293

If the p_value of Leven't test is greater than 0.05, you can assume equal variance. Here 0.0293  is less than 0.05, hence the variance is not equal.

Here the second assumption is failed. To move forward use "equal_var=False" as a parameter when calling ttest_ind function.

When one of the assumption is false, you can use Welch T-Test. You dont have to call this function, because ttest_ind will automatically call Welsh ttest.

Here P-value: 1.4988271248674483e-05
this is equivalent to 0.000014988271248674483

Here the p_value is less than 0.05 hence reject the Null Hypothesis.

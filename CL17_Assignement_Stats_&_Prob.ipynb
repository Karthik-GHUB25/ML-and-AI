- n = 20
- Mean (xÃÑ) ‚âà 60.15
- Standard Deviation (s) ‚âà 20.36
- Degrees of Freedom (df) = 19
- t-critical value for 95% CI and df = 19 ‚âà 2.093 (from t-table)

‚úÖ Step 2: Compute Standard Error (SE)

SE = \frac{s}{\sqrt{n}} = \frac{20.36}{\sqrt{20}} ‚âà 4.55


‚úÖ Step 3: Compute Margin of Error (ME)

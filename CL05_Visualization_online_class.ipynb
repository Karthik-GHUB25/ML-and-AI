# **Histogram**

Step 1: What Is Probability? Before we get to PDFs, letâ€™s start with basic probability:

If you roll a die, the chance of getting a 4 is 1 out of 6 â†’ thatâ€™s a discrete probability.
You can list all possible outcomes: 1, 2, 3, 4, 5, 6. This works because the outcomes are countable and exact.

ðŸŒŠ Step 2: What If the Outcomes Arenâ€™t Countable? Now imagine you're measuring height or MPG:

You canâ€™t list every possible value â€” there are infinite possibilities between 20 and 21 MPG. So instead of asking â€œWhatâ€™s the probability of exactly 30 MPG?â€ (which is basically zero), we ask: â€œWhatâ€™s the probability that MPG falls between 28 and 32?â€ This is where continuous probability comes in â€” and where PDF becomes useful.

ðŸ“˜ Step 3: What Is a PDF?A Probability Density Function is a curve that shows how data is distributed across a continuous range.- It doesnâ€™t give the probability of a single value. It shows how dense the data is around each value.
The area under the curve between two values gives the actual probability.

ðŸ“Š Step 4: PDF Graph â€” What It Looks LikeImagine a smooth hill-shaped curve:- X-axis: The values of your variable (e.g. MPG).

Y-axis: The density â€” how concentrated the data is around each value. If the curve is tall at 30 MPG, it means many cars have MPG near 30.

ðŸ“
Step 5: Area Under the Curve = ProbabilityLetâ€™s say you want to know:â€œWhatâ€™s the probability that a car has MPG between 28 and 32?â€Youâ€™d look at the area under the PDF curve from 28 to 32. That area = the probability.

ðŸ›  Real-Life AnalogyThink of a mountain range:- The height of the mountain at each point = how many hikers are there.

The area between two points = how many hikers are in that zone. The PDF is like that mountain â€” it shows where the data is â€œdenseâ€ or â€œsparse.â€âœ… Summary| Concept | Meaning | | PDF | A curve showing how data is distributed over a continuous range | | X-axis | Values of the variable | | Y-axis | Density (not probability) | | Area under curve | Actual probability between two values | | Total area | Always equals 1 (100% probability) |

The Probability Density Function (PDF) is a foundational concept in statistics and machine learning. Letâ€™s break it down intuitively and technically:


```python
# Comparing the Histogram between matplotlib and seaborn
plt.figure(figsize=(8,3))

#Matplotlib version
plt.subplot(1, 2, 1)                 # 1 row, 2 columns
                                     # plt() = Refers to matplotlib.pyplot, the plotting module. Usually imported as import matplotlib.pyplot as plt.

plt.hist(data['mpg'], bins=10)       # data['mpg'] = The data being visualized. This accesses the 'mpg' column from the DataFrame data.
                                     # hist() = A function that creates a histogram â€” a bar chart showing the distribution of numeric data.
                                     # bins=10 = Divides the range of 'mpg' values into 10 intervals (bins). Each bin counts how many values fall into that range.
plt.title('Matplotlib Histogram')

#Seaborn version
plt.subplot(1, 2, 2)  # 1 row, 2 columns
sns.histplot(data['mpg'], bins=10, kde=True) # Adds a smooth curve that estimates the probability density of the data â€” helps visualize the shape of the distribution.
plt.title('Seaborn Histogram')
plt.show()
```

## **Gaussian Distribution**

```python
# Create a Gaussian Distribution (Normal Bell Curve).  The center of the curve is the mean. Standard deviations of the curve,
# Normal curve, Average value, stadnadard deviation value. using linspace arrange them into 4 equidistant points
#
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

### PDF - Probability Density Function

PDF is what creates the normal curve

Standard deviation is a measure of how spread out the values in a dataset are. It tells you how much the individual data points deviate from the mean (average) of the dataset.


```python
mean, std = 0, 1             # Standard Deviation SD = 1, average = 0. Here SD is 1, which mean between 0 and 4 there is 4 data points equally spread.
x = np.linspace(-4, 4, 100)  # -4 to 4 with 100 equidistant points. There are 100 points on the curve from -4 to 0 and 100 points from 0 to 4
y = norm.pdf(x, mean, std)   # For every values in the point we create Y values using Probability density function

plt.figure(figsize=(4,3))
plt.plot(x, y, label = 'Normal Distribution')
plt.show()
```

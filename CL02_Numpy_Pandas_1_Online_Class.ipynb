```python
# lenght of math and science should be the same

# Comparison of python and numpy

# Goal is to know how much time it is taking. so for that import the time

# using python.
import time
def using_lists():
  # Start time - before start of the operation
  start_time = time.time()

  math = [65, 85, 78, 99]
  science = [68, 84, 72, 90]

  # perform element wise addition using python
  [math[i] + science[i] for i in range(len(math))]

  # end time - after the completion of the operation
  end_time = time.time()

  duration = end_time - start_time
  return duration
# print(duration) # Will give 0.00015997886657714844
```

```python
# using numpy array
def using_numpy_array():
  # Start time - before start of the operation
  start_time = time.time()

  math = np.array([65, 85, 78, 99])
  science = np.array([68, 84, 72, 90])

  student_score = math + science # it performs element wise addition

  # end time - after the completion of the operation
  end_time = time.time()

  a_duration = end_time - start_time
  return a_duration
```

```python
# The number of elements in the above a very little and the processing is super fast.
# lets put 100,000 elements
# Using python
import time
def using_lists_large(n):
  # Start time - before start of the operation
  start_time = time.time()

  list_1 = range(n)
  list_2 = range(n)

  # perform element wise addition using python
  [list_1[i] + list_2[i] for i in range(len(list_1))]

  # end time - after the completion of the operation
  end_time = time.time()

  l_duration = end_time - start_time
  return l_duration
```

```python
# using numpy array
def using_numpy_array_large(n):
  # Start time - before start of the operation
  start_time = time.time()

  array_1 = np.arange(n)
  array_2 = np.arange(n)

  array_total = array_1 + array_2 # it performs element wise addition

  # end time - after the completion of the operation
  end_time = time.time()

  la_duration = end_time - start_time
  return la_duration
```

```python
# Below gives the same result
cars[Condition1 & Condition2]
```

```python
# The above process is time consuming
# Approach 2
Condition1 = (cars['cyl'] > 6)
Condition2 = (cars['wt'] >= 3.8)

#condition = np.logical_and(cars['cyl'] > 6, cars['wt'] >= 3.8)
condition = np.logical_and(Condition1, Condition2)
cars[condition]
```

```python
# Using query
# Approach3. Pandas supports multi-threading
cars.query('cyl > 6 and wt >= 3.8')
```

```python
# Drawback is supported for AND operator and not for OR operator
# Logical OR Gate
Condition1 = (cars['cyl'] > 6)
Condition2 = (cars['wt'] >= 3.8)
condition = np.logical_or(Condition1, Condition2)
cars[condition]
```

```python
# More than two condiion
condition1 = (cars['cyl'] > 6) # AND
condition2 = (cars['wt'] >= 3.8) # AND
condition3 = (cars['hp'] >= 100) # OR
condtion = np.logical_or(np.logical_and(condition1, condition2), condition3)
cars[condition]
```

We should not remove outliers from this data because it is a medical data and the outliers are also crucial datapoint.
When to Keep Outliers
Keep them if:
- Theyre real and meaningful: For example, a patient with extremely high cholesterol might be rare but medically important.
- Youre doing anomaly detection: Outliers are the signal, not noise.
- You want to understand edge cases: In QA or business systems, outliers might reveal bugs, fraud, or system limits.
- Youre modeling rare events: Like predicting system failures or high-value transactions.
Note:
For Decision Tree we dont have to remove the outliers it will take care of itself

When to Remove Outliers
Remove them if:
- Theyre errors or noise: Like a blood pressure of 900 or a negative age.
- They distort your model: Linear regression, for example, is sensitive to extreme values.
- Youre building general-purpose models: Like predicting average customer behavior or typical system performance.
- Youre visualizing data: Outliers can skew plots and hide patterns.


```python
# Outlier List. The box plot will show you the outlier
# Out_list = ["restingBloodPressure",  "cholesterol"]
#for i in Out_list:
#  Q1 = df[i].quantile(0.25)
#  Q3 = df[i].quantile(0.75)
#  IQR = Q3 - Q1
#  lower_bound = Q1 - 1.5 * IQR
#  df = df[(df[i] >= lower_bound) & (df[i] <= upper_bound)]
```

```python
#Below code will show the outliers in all the columms
def detect_outliers_iqr(df, columns):
    outlier_report = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Find outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_report[col] = {
            "outlier_count": outliers.shape[0],
            "outlier_rows": outliers.index.tolist()
        }

    return outlier_report

# Example usage
columns_to_check = ['age', 'gender', 'chestPain', 'restingBloodPressure', 'cholesterol',
                    'fastingBloodSugar', 'RestingECGresults', 'maximumHeartRate',
                    'exerciseInducedAngina', 'oldpeak', 'slope', 'numberOfMajorVessels',
                    'thalassemia', 'target']

report = detect_outliers_iqr(df, columns_to_check)

# Print summary
for col, info in report.items():
    if info["outlier_count"] > 0:
        print(f"{col}: {info['outlier_count']} outliers found at rows {info['outlier_rows']}")
    else:
        print(f"{col}: No outliers detected")
```

#13. Depth

How to restrict the depth? or How many times to want to split?

Hyperparameter tuning

Here you can give the number until you are satisfied.

```python
depths = np.arange(1, 13)
for i in depths:
    model = DecisionTreeClassifier(max_depth=i, random_state=60)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy for depth {i}: {acc}")
#
```

Here from depth 8th it is all the same result. it is advisable to stop at early like at depth 5.

small (1 - 4) = to capture simple pattern

Medium (5-10) = often perform the best

Large (15+) = May overfit

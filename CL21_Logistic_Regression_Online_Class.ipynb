```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function definition
def sigmoid(x, k=1, x0=0, L=1):
    return L / (1 + np.exp(-k * (x - x0)))

# Generate x values
x = np.linspace(-10, 10, 500)
y = sigmoid(x, k=1, x0=0, L=1)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(x, y, label=r'$f(x) = \frac{1}{1 + e^{-k(x - x_0)}}$', color='blue')
plt.axhline(0.5, color='gray', linestyle='--', label='Decision Threshold (0.5)')
plt.axvline(0, color='red', linestyle='--', label='Midpoint $x_0$')
plt.title('Logistic Function (Sigmoid Curve)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

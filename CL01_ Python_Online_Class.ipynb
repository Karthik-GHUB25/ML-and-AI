```python
# Question 5:
# Check if the number if prime number.
# A prime number has only two divisors: 1 and itself with remainder 0
def is_prime(n):
  if n <= 1:
     print("Not a prime")
     return False
  for i in range(2,n):
    if n%i == 0:  #If n is divisible by any i from 2 to n - 1, it’s not a prime.
      return False
  return True
  print("It is a prime")

print(is_prime(1))
print(is_prime(2))
print(is_prime(3))
print(is_prime(4)) # n%i = (4%2) (4%3) (4%4) to check all possible divisors from 2 up to n - 1 to ensure that no number can divide n without a remainder.
print(is_prime(5))
```

```python
# Question 12:
# Write a code to print Fibonacci sequence up to n terms
def fibonacci(n):
    if n <= 0:
        print("Please enter a positive integer.")
        return
    fib_seq = [0, 1]  # Start with the first two numbers
    for i in range(2, n):  # Generate sequence up to n terms
        fib_seq.append(fib_seq[-1] + fib_seq[-2])

    print("Fibonacci sequence:", fib_seq[:n])

# Example usage
num_terms = int(input("Enter number of terms: "))
fibonacci(num_terms)
```

```python
# To find duplicates in the list
def find_duplicates(input_list):
    seen = set()   # here are seen and duplicates are two empty sets
    duplicates = set()
    for item in input_list: # For every element that is read in the list we are checking if is in seen, if yes then it written in duplicates
        if item in seen:
            duplicates.add(item)
        else:
            seen.add(item)
    return list(duplicates)

exer_li = [10, 'ABC', 20, 30, 40, '42x', 30, 951, 1000, 'ABC', 40]
print(find_duplicates(exer_li))  # Output: ['ABC', 40, 30]
```

```python
# The above function can be done through list
def find_ordered_duplicates(input_list):
    seen = []       # Here seen and duplicates are list
    duplicates = []
    for item in input_list:
        if item in seen and item not in duplicates:
            duplicates.append(item)
        else:
            seen.append(item)
    return duplicates

exer_li = [10, 'ABC', 20, 30, 40, '42x', 30, 951, 1000, 'ABC', 40]
print(find_ordered_duplicates(exer_li)) # Difference would be ordered collection whereas the above it gives unordered collection
```

```python
# ################################## Hybrid Inheritence ##################################
# It is nothing but multiple and multi-level inheritance
class GrandFather:
  def receipe(self):
    print('This is our traditional food receipe')

  def property(self):
    print('This is our village property')

class Father (GrandFather): # This father class is inheriting Grand Father class - Child Class
  def car(self):
    print('This is our Car')

  def House(self):
    print('This is our House')

class Uncle(GrandFather): # This Uncle class is inheriting Grand Father class = Child Class
  def home(self):
    print('This is our Home')

class Son (Father, Uncle): # This son class inheriting from Uncle and Father Class - Multiple Inheritance - Child Class
  def phone(self):
    print('This is my Phone')

  def bike(self):
    print('This is my bike')

use_obect_20 = Son()
use_obect_20.bike()  # Son Class Function
use_obect_20.car()   # Father Class Function
use_obect_20.receipe() # Grand Father Class Function
```

```python
################################################################################################
###################################### Polymorphism ###########################################

# Poly -> Many
# Morphism -> Forms
#
# Method Overloading : In a file we can create more than one function with same name but number of parameter must be different
# Method Overriding

# Method Overloading is not supported in Python

def sum(a, b):
  print(a+b)

def sum(a, b, c): # Interperter will override sum(a, b). It will always execute the latest one
  print(a+b+c)

sum(10,20)   # This will throw error, Interperter remember the last sum function (def sum(a, b, c):)
sum(10,20, 30) # This will override the previous function.

# Method overriding: It will execute the latest function and only consider that function.
# It will override the previous function and will always consider the latest function

# Method overloading also know as complile time polymorphism
# Method overriding also known as runtime ploymorphism
```

```python
# Method overriding works
class user:  # Parent class
  def registered_details(self, name, age):
    self.name = name
    self.age = age

  def course_access(self):
    print('You can access the course here')

class Admin (user):    #Child class 1
  def Agenda(self):
    print('You can access course here')

  def team_update(self):
    print('Need to check for updates')

  def course_access(self): # Here we are overriding the parent class course_access function
    print('We can add courses to learners and remove courses from learners')

user_object_21 = Admin()
user_object_21.course_access()

user_object_22 = user()
user_object_22.course_access() # During the run time interperter does not know there is a child class
```

```python
#################################### 07/04/2025 ##################################################
################################### Encapsulation ####################################

# Protecting data/Properties/functions from outside world

class Bank:
  def withdraw(self):
    print(' You can withdraw money here')

  def Deposit(self):
    print('You can deposit money here')

  def ATM(self):
    print('You can access ATM here')

# We need to protect Locker function. This function should restricted
  def __Locker(self): #__ will make the function Private. This is what is called Encapsulation.
    print('You can access the locker here')
# When you create a private function it cannot be accessed outside the class

# Create a new function Manager and try to access this Locker function (Private function) inside it
# Local function are accessed through a function
  def Manager(self):
    inp = int(input(' Enter your Account Number :'))
    print('Thanks for providing the details')
    self.__Locker() # This is how you access private function

user_object_23 = Bank()
user_object_23.withdraw()
user_object_23.Manager()
user_object_23.Locker() # This will throw error. We cannot access private function directly
```

```python
####################################### Abstraction ##########################################
# Showing only neccessary information to the user and hiding the implementation
# A practical example would be driving a car. A driving school will teach only on how to drive a car, and would not teach any mechanics of car.
# Sharing the driving instruction is beneficial to the driver than sharing Mechanics of the car.

# Step 1, import ABC and abstractmethod from abc
#         ABC -> help us to create abstract class. abstractmethod -> will help us to create abstract method
# Step 2, import ABC Module Abstract Base Classes

from abc import ABC, abstractmethod #
import abc
# When we create an abstract class, we will create all the function as abstract method

class Vehicle(ABC): # ABC is inherited, this is also an abstract class
                    # Every Abstract class contain Abstract Method
                    # Parent Class set the rules. Which means Child class must have all the below method apart from having its own method
  @abstractmethod
  def Start_Engine(self): # we are not going to implement function. Function will be implemented via child class
    pass # We use pass keyword to create empty functions

  @abstractmethod  # A decorator which help us create abstract method. It is like a representation of abstract method
  def Drive(self):
    pass

  @abstractmethod
  def Stop_Engine(self):
    pass

  @abstractmethod
  def Park(self):
    pass

# Until this point four abstract methods are created and not impletmented
# Any child class inherit the Vehicle class will be implementing those 4 methods
# When this shown to someone or someone looking into the code, they will not see the implementation
# Any number of class can inherit Vehicle class

```

```python
class Car(Vehicle): # here Vehicle class is inherited. This is a child class where we do the implementation
  # Below method has to be present, because it is a blue print from the parent class.
  def Start_Engine(self):
    print('Car Engine Started')

  def Drive(self):
    print('Car is running')

  def Stop_Engine(self):
    print('Car Engine Stopped')

  def Park(self):
    print('Car is in the park Mode')

  def Specs(self): # This is additional method that is apart from mandatory methods.
    print('This is our car specs')

user_object_27 = Car()
user_object_27.Start_Engine()
user_object_27.Drive()
user_object_27.Stop_Engine()
user_object_27.Park()
user_object_27.Specs()
```

```python
# Question 3:
# Create a class named ‘Encapsulation’:
# a. Inside the class, first create a constructor. Inside the constructor, initialize originalValue variable as 10.
# b. After creating the constructor, define a function named ‘Value’ and # this function should return the variable that we have initialized in the constructor.
# c. Now create 2nd function named setValue, and pass an argument named ‘newValue’.
# The task of this function will be to replace the value of the originalValue variable by the value of the newValue variable.

class Encapsulation:
  def __init__(self):
    self.originalValue = 10

  def Value(self):
    return self.originalValue

  def setValue(self, newValue):
    self.originalValue = newValue

  def printValue(self):
    print(self.originalValue)

user_object_32 = Encapsulation()
user_object_32.printValue()
user_object_32.setValue(1000)
user_object_32.printValue()
```

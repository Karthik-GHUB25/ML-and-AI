This means 46% of the economy coming from service sector.

```python
# Visualize
sns.histplot(df['Economy: Services and other activity (% of GVA)'], color = 'Salmon', kde = True)
plt.title('Services Sector')
plt.xlabel('% of GVA')
plt.ylabel('Frequency')
plt.show()
```

What the Graph Shows

• 	X-axis (% of GVA): This represents how much of a country's Gross Value Added comes from the services sector.

• 	Y-axis (Frequency): How many countries fall into each percentage range.

• 	Main concentration: Most countries fall between 40% and 80%, with a peak around 60%.

• 	Outlier: There’s a single bar at –100%, which is likely a data error or placeholder.


It is a negatively skewed.

KDE stands for Kernel Density Estimate. When you add kde=True in sns.histplot(), it overlays a smooth curve that estimates the distribution of your data. It’s great for:
- Seeing the shape of the distribution more clearly
- Spotting skewness, modality, or outliers
- Comparing multiple distributions visually
When to Use KDE
✅Use it when:
- You want a smooth visual summary of the data
- You're working with continuous variables
- You're exploring patterns or anomalies (like in QA or anomaly detection)

❌ When to Skip KDE
Skip it when:
- Your data is discrete (e.g., counts, categories, survey responses)
- You want to emphasize actual frequencies rather than smoothed estimates
- You're doing formal statistical analysis where raw histograms are more appropriate


```python
# Box plot
plt.boxplot(df['Economy: Services and other activity (% of GVA)'],
patch_artist = True,
boxprops = dict(facecolor = 'lime'))
plt.title('Services Sector')
plt.xlabel('% of GVA')
plt.ylabel('Frequency')
plt.show()
```

🔍 Why the Box Plot Looks Narrower
Your histogram showed a wide spread of values between 40% and 80%, with a peak around 60%.
But the box plot focuses on:
- Q1 (25th percentile): ~50%
- Q3 (75th percentile): ~60%
- So the IQR (Interquartile Range) = 60% − 50% = 10%
This means:
The middle 50% of countries have service-sector contributions between 50% and 60%.

That’s a tight cluster, even though the full range (from histogram) is wider.

Select random sample from the dataset and compare it with population mean

```python
import numpy as np

# Step 1: Set a random seed for reproducibility (optional)
# np.random.seed(42) ---> This is equivalent to random_state.
# On the step 2 just use sample_df = df.sample(n=100)

# Step 2: Select a random sample of 100 rows
sample_size = 100
sample_df = df.sample(n=sample_size, random_state = 1)  # random_state = 1 equivalent to Seed

# Step 3: Calculate the mean population of the sample
sample_mean = sample_df['Population in thousands (2017)'].mean()

# Step 4: Calculate the mean population of the entire dataset
population_mean = df['Population in thousands (2017)'].mean()

# Step 5: Compare the two means
print(f"Sample Mean Population: {sample_mean:,.2f}")
print(f"Population Mean: {population_mean:,.2f}")
print(f"Difference: {sample_mean - population_mean:,.2f}")
```

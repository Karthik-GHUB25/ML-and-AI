```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load diamonds dataset
diamonds = sns.load_dataset('diamonds')

#  Correlation heatmap (numerical features only)
plt.figure(figsize=(10, 6))
sns.heatmap(diamonds.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation of Numerical Features with Price')
plt.tight_layout()
plt.show()

#  Scatter plot: Carat vs Price
sns.scatterplot(x='carat', y='price', data=diamonds, alpha=0.5)
plt.title('Carat vs Price')
plt.show()

#  Box plot: Cut vs Price
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Cut Quality vs Price')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Sample data
X = np.array([1, 2, 3, 4])
Y = np.array([2, 4, 5, 7])

# Grid of M and C values
M_vals = np.linspace(0, 3, 100)
C_vals = np.linspace(0, 3, 100)
M_grid, C_grid = np.meshgrid(M_vals, C_vals)

# Compute SSE for each pair
SSE = np.array([
    np.sum((Y - (m * X + c))**2)
    for m, c in zip(np.ravel(M_grid), np.ravel(C_grid))
]).reshape(M_grid.shape)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(M_grid, C_grid, SSE, cmap='viridis')
ax.set_xlabel('Slope (M)')
ax.set_ylabel('Intercept (C)')
ax.set_zlabel('SSE')
plt.show()
```

```python
# Given a sample list, tuple, and set with elements from 1 to 10
# Convert each of the data types into a pandas series
# Create a separate list of index values for the series

import pandas as pd
# Sample data from 1 to 10
sample_list = list(range(1, 11))
sample_tuple = tuple(range(1, 11))
sample_set = set(range(1, 11))

print(sample_list)
print(sample_tuple)
print(sample_set)
```

```python
#A Pandas Series is like a supercharged one-dimensional array — but with labels

ind = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
# If you don’t define an index when creating a Pandas Series, it automatically assigns a default integer index starting from 0 and incrementing by 1.


# Convert to pandas Series
lst_ser = pd.Series(sample_list, index=ind)
print(lst_ser)

# Convert to pandas Series
tup_ser=pd.Series(sample_tuple, index=ind)
print(tup_ser)

# Convert set to series
# Convert set to list
set_ser=pd.Series(list(sample_set), index=ind)
print(set_ser)
```

```python
# Question 4: Create a sample pandas series with x integer elements. Perform the following operations
# A. Print the first 5 elements of the series
# B. Print the last 5 elements of the series
# C. Insert an integer value 50 at the inder 5
# D. Update the last element in the series with the value 100
# E. Delete the first elelment from the series

pand_ser1 = pd.Series(range(11, 21))
print(pand_ser1)
```

```python
#A. Print the first 5 elements of the series
print(pand_ser1.head(5))
```

```python
# B. Print the last 5 elements of the series
print(pand_ser1.tail(5))
```

```python
# C. Insert an integer value 50 at the index 5

ind_pos = 5
val = 50
# Rebuild and reset index
pand_ser1 = pd.concat([
    pand_ser1.iloc[:ind_pos],
    pd.Series([val]),
    pand_ser1.iloc[ind_pos:]
]).reset_index(drop=True)

print(pand_ser1)
```

```python
# D. Update the last element in the series with the value 100
pand_ser1.iloc[-1] = 100
print(pand_ser1)
```

```python
# E. Delete the first elelment from the series
pand_ser1 = pand_ser1.iloc[1:]
# Or
# pand_ser1 = pand_ser1.drop(0)
# or
# pand_ser1 = pand_ser1.reset_index(drop=True) # This will reset the index from zero
print(pand_ser1)
```

```python
# Question 5: For the given series A and B, perform concatenation on the same series,
# series_A=pd.Series([1, 2, 3, 4, 5]), series_B=pd.Series([6, 7, 8, 9, 10])
import pandas as pd

ConCat_Ser = pd.concat(
    [pd.Series([1, 2, 3, 4, 5]), pd.Series([6, 7, 8, 9, 10])],
    ignore_index=True) # Correctly placed inside pd.concat()


print(ConCat_Ser)
```

```python
# Without ignore_index
ConCat_Ser = pd.concat(
    [pd.Series([1, 2, 3, 4, 5]), pd.Series([6, 7, 8, 9, 10])])
print(ConCat_Ser)
# Index will not reset
```

```python
# Question: 7 - Create a dataFrame with two columns ['Name', 'Age']. Perform the following operations
# A. Print the first 3 rowsof the DataFrame
# B. Print the rows that have age greater than 15.
# C Insert 5 imaginary names and age to the padas dataframe and merge it with the previous dataframe
# D. Update the 5th row so that the name = John and age = 56
# E. Delete the first row from the dataframe


# Create the DataFrame with two columns
DF_Nam_Age1 = pd.DataFrame({
    'Name': ['Tom Cruise', 'Brad Pitt', 'Tom Hanks', 'Johnny Depp', 'Leonardo DiCaprio', 'Robert Downey Jr.', 'Jerry Seinfeld'],
    'Age': [61, 60, 67, 61, 48, 58, 69]
})

print(DF_Nam_Age1)
```

```python
# A. Print the first 3 rowsof the DataFrame
print(DF_Nam_Age1.head(3))
```

```python
# B. Print the rows that have age greater than 15.
print(DF_Nam_Age1[DF_Nam_Age1['Age'] > 15])
```

```python
# C Insert 5 imaginary names and age to the pandas dataframe and merge it with the previous dataframe

DF_Nam_Age2 = pd.DataFrame({
    'Name': ['Lewis Koumas', 'Prince George', 'Like Nastya', 'Mille Bobby Brown', 'Erling Haaland'],
    'Age': [19, 12, 11, 20, 25]
})

DF_Nam_Age3 = pd.concat([DF_Nam_Age1, DF_Nam_Age2], ignore_index=True)


print(DF_Nam_Age3)
```

```python
# D. Update the 5th row so that the name = John Doe and age = 56

DF_Nam_Age3.loc[4, 'Name'] = 'John Doe'
DF_Nam_Age3.loc[4, 'Age'] = 56

print(DF_Nam_Age3)
```

```python
# E. Delete the first row from the dataframe
DF_Nam_Age3 = DF_Nam_Age3.drop(0)
print(DF_Nam_Age3)
```

```python
# Question 8: Create a DataFrame with columns Name and Age. Perfrom the following Operation:
# A. Write a sample code to search for the Name where Age is 20.
# B. Sort the Data Frame in Ascending order of ages.
# C. Using python loops,print the ages from the Dataframe and store it in a set

DF_Nam_Age4 = pd.concat([DF_Nam_Age1, DF_Nam_Age2], ignore_index=True)
print(DF_Nam_Age4)
```

```python
# A. Write a sample code to search for the Name where Age is 20.
print(DF_Nam_Age4[DF_Nam_Age4['Age'] == 20])
#
```

```python
# B. Sort the Data Frame in Ascending order of ages.
DF_Nam_Age4 = DF_Nam_Age4.sort_values(by='Age')
print(DF_Nam_Age4)
```

```python
# C. Using python loops,print the ages from the Dataframe and store it in a set
age_set = set()
for age in DF_Nam_Age4['Age']:
    age_set.add(age)
print(age_set)
```
